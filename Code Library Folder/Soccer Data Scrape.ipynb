## Soccer Data Web Scrapping

#### Importing Required libraries

```python
#importing required libraries
import pandas as pd
import numpy as np

import re
import requests

from bs4 import BeautifulSoup as Soup
# operator.le used to perform less than or greater than operation
from operator import le

import warnings
warnings.simplefilter(action = 'ignore', category=Warning)
```

#### Data Collection from sofifa.com

```python
#Scraping soccer data from sofifa.com
base_url = 'https://sofifa.com'
url = 'https://sofifa.com/players?offset='
column = ['ID', 'Name', 'Position', 'Age', 'Overall', 'Potential', 'Team',
          'Value', 'Wage', 'Total_Point', 'height', 'weight', 'crossing', 'finishing', 
          'heading_accuracy', 'short_passing', 'volleys',
          'acceleration', 'sprint_speed', 'agility', 'reactions', 'balance', 'dribbling', 'curve',
          'fk_accuracy', 'long_passing', 'ball_control', 'shot_power', 'jumping', 'stamina',
          'strength', 'long_shots', 'aggression', 'interception', 'positioning', 'vision', 'penalties',
          'composure', 'defensive_awareness', 'standing_tackle', 'sliding_tackle', 'gk_diving',
          'gk_handling', 'gk_kicking', 'gk_positioning', 'gk_reflexes', 'preferred_foot', 'weak_foot',
          'skill_move', 'international_reputation', 'work_rate', 'body_type']
print('The length of the column is', len(column))
df_fifa = pd.DataFrame(columns = column)
```

```python
#Creating function to get player url.
def get_soup(player_url):
  soup = requests.get(player_url)
  soup = Soup(soup.text, 'html.parser')
  #getting height and weight of player
  height_weight = soup.select_one('.meta.ellipsis').text.split(' ')
  #height of player
  height = height_weight[-2]
  #weight of player
  weight = height_weight[-1]

  #attacking
  crossing = ''
  finishing = ''
  heading_accuracy = ''
  short_passing = ''
  volleys = ''

  #skill
  dribbling = ''
  curve = ''
  fk_accuracy = ''
  long_passing = ''
  ball_control = ''

  #movement
  acceleration = ''
  sprint_speed = ''
  agility = ''
  reactions = ''
  balance = ''

  #power
  shot_power = ''
  jumping = ''
  stamina = ''
  strength = ''
  long_shots = ''

  #mentality
  aggression = ''
  interception = ''
  positioning = ''
  vision = ''
  penalties = ''
  composure = ''

  #defender
  defensive_awareness = ''
  standing_tackle = ''
  sliding_tackle = ''

  #goalkeeper
  gk_diving = ''
  gk_handling = ''
  gk_kicking = ''
  gk_positioning = ''
  gk_reflexes = ''

  #profile
  preferred_foot = ''
  weak_foot = ''
  skill_move = ''
  international_reputation = ''
  work_rate = ''
  body_type = ''

  
  block_quarter = soup.select('.block-quarter')
  for quarter in block_quarter:
    try:
      attr = quarter.select_one('h5')
      #The max an ability can go is 99
      if attr.text == 'Attacking':
        ul = quarter.select_one('ul')
        lis = ul.select('li')
        #player crossing ability
        crossing = lis[0].select_one('span.bp3-tag').text
        #player finishing ability
        finishing = lis[1].select_one('span.bp3-tag').text
        #player heading accuracy ability
        heading_accuracy = lis[2].select_one('span.bp3-tag').text
        #player short_passing ability
        short_passing = lis[3].select_one('span.bp3-tag').text
        #player volleys(kicking the ball in the air with accuracy before it touches the floor) ability
        volleys = lis[4].select_one('span.bp3-tag').text

      if attr.text == 'Movement':
        ul = quarter.select_one('ul')
        lis = ul.select('li')
        #player acceleration ability
        acceleration = lis[0].select_one('span.bp3-tag').text
        #player sprint_speed ability
        sprint_speed = lis[1].select_one('span.bp3-tag').text
        #player agility ability
        agility = lis[2].select_one('span.bp3-tag').text
        #player reactions ability
        reactions = lis[3].select_one('span.bp3-tag').text
        #player balance ability
        balance = lis[4].select_one('span.bp3-tag').text

      if attr.text == 'Skill':
        ul = quarter.select_one('ul')
        lis = ul.select('li')
        #player dribbling ability
        dribbling = lis[0].select_one('span.bp3-tag').text
        #player curve ability
        curve = lis[1].select_one('span.bp3-tag').text
        #player fk_accuracy ability
        fk_accuracy = lis[2].select_one('span.bp3-tag').text
        #player long_passing ability
        long_passing = lis[3].select_one('span.bp3-tag').text
        #player ball_control ability
        ball_control = lis[4].select_one('span.bp3-tag').text

      if attr.text == 'Power':
        ul = quarter.select_one('ul')
        lis = ul.select('li')
        #player shot_power ability
        shot_power = lis[0].select_one('span.bp3-tag').text
        #player jumping ability
        jumping = lis[1].select_one('span.bp3-tag').text
        #player stamina ability
        stamina = lis[2].select_one('span.bp3-tag').text
        #player strength ability
        strength = lis[3].select_one('span.bp3-tag').text
        #player long_shots ability
        long_shots = lis[4].select_one('span.bp3-tag').text

      if attr.text == 'Mentality':
        ul = quarter.select_one('ul')
        lis = ul.select('li')
        #player aggression ability
        aggression = lis[0].select_one('span.bp3-tag').text
        #player interception ability
        interception = lis[1].select_one('span.bp3-tag').text
        #player positioning ability
        positioning = lis[2].select_one('span.bp3-tag').text
        #player vision ability
        vision = lis[3].select_one('span.bp3-tag').text
        #player penalties ability
        penalties = lis[4].select_one('span.bp3-tag').text
        #player composure ability
        composure = lis[5].select_one('span.bp3-tag').text

      if attr.text == 'Defending':
        ul = quarter.select_one('ul')
        lis = ul.select('li')
        #player defensive_awareness ability
        defensive_awareness = lis[0].select_one('span.bp3-tag').text
        #player standing_tackle ability
        standing_tackle = lis[1].select_one('span.bp3-tag').text
        #player sliding_tackle ability
        sliding_tackle = lis[2].select_one('span.bp3-tag').text

      if attr.text == 'Goalkeeping':
        ul = quarter.select_one('ul')
        lis = ul.select('li')
        #player gk_diving ability
        gk_diving = lis[0].select_one('span.bp3-tag').text
        #player gk_handling ability
        gk_handling = lis[1].select_one('span.bp3-tag').text
        #player gk_kicking ability
        gk_kicking = lis[2].select_one('span.bp3-tag').text
        #player gk_positioning ability
        gk_positioning = lis[3].select_one('span.bp3-tag').text
        #player gk_reflexes ability
        gk_reflexes = lis[4].select_one('span.bp3-tag').text

      if attr.text == 'Profile':
        ul = quarter.select_one('ul')
        lis = ul.select('li')
        #player preferred foot
        preferred_foot = lis[0].text
        #player weak foot rating (1-5)
        weak_foot = lis[1].text
        #player skill move rating (1-5)
        skill_move = lis[2].text
        #player international reputation rating
        international_reputation = lis[3].text
        #player work rate
        work_rate = lis[4].text
        #player body_type
        body_type = lis[5].text

    except:
      continue
  return (height, weight, crossing, finishing, heading_accuracy, short_passing, volleys,
          acceleration, sprint_speed, agility, reactions, balance, dribbling, curve,
          fk_accuracy, long_passing, ball_control, shot_power, jumping, stamina,
          strength, long_shots, aggression, interception, positioning, vision, penalties,
          composure, defensive_awareness, standing_tackle, sliding_tackle, gk_diving,
          gk_handling, gk_kicking, gk_positioning, gk_reflexes, preferred_foot, weak_foot,
          skill_move, international_reputation, work_rate, body_type)

#Using beautifulsoup with function above to scrape the soccer data.

#setting limit of data to scrape. We want to scrape the max amount of players which is 20,000.
#60 players on each page so 20000/60 = 333.333 so we have approx max of 333 pages.

#Note this will probably get a few players as duplicates as it is not accurate.
offset = 0
for offset in range(0,335):
  url = url + str(offset*60)
  #print(url)
  
  request = requests.get(url)
  soup = request.text

  data = Soup(soup, 'html.parser')
  table = data.find('tbody')

  for stats in table.findAll('tr'):
    td = stats.findAll('td')

    player_url = base_url + td[1].find('a')['href']
    player_data = get_soup(player_url)

    ID = td[0].find('img').get('id')
    Name = td[1].findAll('a')[0].text
    Position = td[1].findAll('a')[1].text
    Age = td[2].text
    Overall = td[3].find('span').text
    Potential = td[4].find('span').text
    Team = td[5].find('a').text
    Value = td[6].text.strip()
    Wage = td[7].text.strip()
    Total_Point = td[8].text.strip()

    height = player_data[0]
    weight = player_data[1]
    crossing = player_data[2]
    finishing = player_data[3] 
    heading_accuracy = player_data[4] 
    short_passing = player_data[5]
    volleys = player_data[6]
    acceleration = player_data[7]
    sprint_speed = player_data[8]
    agility = player_data[9]
    reactions = player_data[10]
    balance = player_data[11]
    dribbling = player_data[12]
    curve = player_data[13]
    fk_accuracy = player_data[14]
    long_passing = player_data[15]
    ball_control = player_data[16]
    shot_power = player_data[17]
    jumping = player_data[18]
    stamina = player_data[19]
    strength = player_data[20]
    long_shots = player_data[21] 
    aggression = player_data[22] 
    interception = player_data[23] 
    positioning = player_data[24]
    vision = player_data[25]
    penalties = player_data[26]
    composure = player_data[27] 
    defensive_awareness = player_data[28]
    standing_tackle = player_data[29]
    sliding_tackle = player_data[30]
    gk_diving = player_data[31]
    gk_handling = player_data[32]
    gk_kicking = player_data[33]
    gk_positioning = player_data[34]
    gk_reflexes = player_data[35]
    preferred_foot = player_data[36] 
    weak_foot = player_data[37]
    skill_move = player_data[38]
    international_reputation = player_data[39]
    work_rate = player_data[40]
    body_type = player_data[41]

    #adding data in dataframe
    df_players_data = pd.DataFrame([[
          ID, Name, Position, Age, Overall, Potential, Team, Value, Wage, Total_Point,
          height, weight, crossing, finishing, heading_accuracy, short_passing, volleys,
          acceleration, sprint_speed, agility, reactions, balance, dribbling, curve,
          fk_accuracy, long_passing, ball_control, shot_power, jumping, stamina,
          strength, long_shots, aggression, interception, positioning, vision, penalties,
          composure, defensive_awareness, standing_tackle, sliding_tackle, gk_diving,
          gk_handling, gk_kicking, gk_positioning, gk_reflexes, preferred_foot, weak_foot,
          skill_move, international_reputation, work_rate, body_type
    ]])
    print(df_players_data)
    df_players_data.columns = column
    df_fifa = df_fifa.append(df_players_data, ignore_index = True)
  url = 'https://sofifa.com/players?offset='

```

#### Writing scrape data to csv file.

```python
#Saving new clean dataset as csv
df_fifa.to_csv('fifa_data.csv', index = None)
print(f'\nRow Count: {df_fifa.shape[0]}')
```
